creating tensors

method1. creating tensor using empty

```python
import torch
a=torch.empty([2,3])
print(a)

'''It creates tensor as empty space and did'nt assign any values just to create tensor memory'''
```

method 2. using zeroes

```python
b=torch.zeros([2,3])
print(b)
'''create a tensor with all values a s0'''
```

method3. using ones

```python
c=torch.ones([2,3])
print(c)

'''create a tensor of all values as 1'''
```

method 4 : using random

```python
d=torch.rand([2,3])
print(d)

'''create tensor with randome values between 0 and 1'''
```

method 5: using mannual seed if want to use the randome tensor again and again

```python
torch.manual_seed(100)
e=torch.rand([2,3])
print(e)
```

method 6 using tensor to customise and make own tensors

```python
torch.tensor([[1,2,3],[4,5,6]])
```

method 7 using arang,linespace.eye

```python
torch.arange(1,10)
```

```python
torch.linspace(0,10,10)
```
